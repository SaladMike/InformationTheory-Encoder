```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_paths = ['75bcjr_dec.csv', '75hard_viterbi.csv', '75soft_viterbi.csv']

# Loa
```

```python
# Load data from files
dataframes = [pd.read_csv(file_path) for file_path in file_paths]

# Extract Noise Power (dB) and Bit Error Rate (BER) from each dataset
noise_power = [df.iloc[0, 1:].values for df in dataframes]
ber = [df.iloc[1, 1:].values for df in dataframes]
labels = ['75bcjr_dec', '75hard_viterbi', '75soft_viterbi']
```

```python
# Plotting
plt.figure(figsize=(10, 6))
for i in range(len(dataframes)):
    plt.plot(noise_power[i], ber[i], label=labels[i], marker='o')

plt.xlabel('Noise Power (dB)')
plt.ylabel('Bit Error Rate')
plt.title('Noise Power vs Bit Error Rate for Different Decoding Methods')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Assuming BER is usually plotted on a log scale
plt.show()
```
